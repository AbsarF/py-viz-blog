```python
import pandas as pd
from altair import *

from palettable import colorbrewer as cb  
```

## Thing 1: Line Chart
***

```python
df = pd.read_csv('ts.csv')
df.head()
```

```python
c = Chart(df).mark_line().encode(
    x='dt',
    y='value',
    color='kind'
)
c
```

```python
c = Chart(df).mark_line().encode(
    x='dt:T',
    y='value',
    color='kind'
)
c
```

```python
c = Chart(df.assign(dt = pd.to_datetime(df.dt))).mark_line().encode(
    x='dt',
    y='value',
    color='kind'
)
c
```

```python
c = Chart(df.assign(dt = pd.to_datetime(df.dt))).mark_line().encode(
    x='dt',
    y='value',
    color=Color('kind', scale=Scale(range=["#914947", "#ebcecb", "#a50245", "#f39450"]))
)
c
```
