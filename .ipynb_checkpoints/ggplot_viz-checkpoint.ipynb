```python
%matplotlib inline

import pandas as pd
from ggplot import *
```

## Thing 1: Line Chart
***

```python
df = pd.read_csv('ts.csv')
df.head()
```

```python
ggplot(df.assign(dt=pd.to_datetime(df.dt)), aes(x='dt', y='value', color='kind')) + \
    geom_line(size=1.5) + \
    xlab('Date') + \
    ylab('Value') + \
    ggtitle('Title')
```
